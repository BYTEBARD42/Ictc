```python
from ai4bharat.transliteration import XlitEngine
```

```python
import pandas as pd

df = pd.read_csv('train.csv')
df.head(1)
```

```python
e = XlitEngine("hi", beam_width=1, src_script_type = "en")
```

```python
def convert_to_native(text):
    out = e.translit_sentence(text)
    return out['hi']
```

```python
from tqdm import tqdm
tqdm.pandas()
```

```python
final_df = pd.DataFrame()
final_df['text'] = df['hin']
final_df['native_text'] = final_df['text'].progress_map(convert_to_native)
```

```python
final_df.head(5)
```

```python
final_df.to_csv("hindi.csv", index=False)
```
